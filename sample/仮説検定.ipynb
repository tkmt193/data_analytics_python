## 点推定と区間推定

```python
import statsmodels.stats.api as sm
```

```python
# 母平均の点推定，母分散の点推定，母平均の95%信頼区間を計算
result = sm.DescrStatsW(df)
print("母平均の推定値：", result.mean)
print("母分散の推定値（不偏分散）：", result.var)
print("95%信頼区間: ", result.tconfint_mean(alpha=0.05))
```

## 仮説検定

```python
# 母平均100gと仮定して仮説検定（ttest_mean(value=100.0)）
t, p, d = result.ttest_mean(value=100.0)
print("p値：",p)
# p(0.166)>a(0.05)となるため、帰無仮説を棄却できない(=母平均は100gでないとはいえない）
```
